Lets try to calculate this distance(or difference between TotalDistance and TrackerDistance)

```python
# Compute residuals
residuals = df['TotalDistance'] - df['TrackerDistance']

# Total number of rows
total_rows = len(residuals)

# Count of zero residuals
zero_residual_count = (residuals == 0).sum()

# Count of non-zero residuals
nonzero_residual_count = total_rows - zero_residual_count

# Calculate percentages
zero_percent = (zero_residual_count / total_rows) * 100
nonzero_percent = (nonzero_residual_count / total_rows) * 100

# Print results
print(f"Rows with zero residuals: {zero_residual_count} ({zero_percent:.2f}%)")
print(f"Rows with non-zero residuals: {nonzero_residual_count} ({nonzero_percent:.2f}%)")
```

```python
(df[['TotalDistance', 'TrackerDistance']]
 .assign(
     residuals = lambda d : (d['TotalDistance'] - d['TrackerDistance'])
 )
 .query('residuals > 0')
 )
```

Lets try to understand the relation between minutesCaloriesNarrow and minutesCaloriesWide

```python
import pandas as pd

# Convert to datetime
df1['ActivityMinute'] = pd.to_datetime(df1['ActivityMinute'])
df2['ActivityHour'] = pd.to_datetime(df2['ActivityHour'])

# Floor df1 minute to hour
df1['ActivityHour'] = df1['ActivityMinute'].dt.floor('H')

# Get sets of unique (Id, Hour) pairs from both datasets
df1_pairs = set(zip(df1['Id'], df1['ActivityHour']))
df2_pairs = set(zip(df2['Id'], df2['ActivityHour']))

# Calculate intersection and overlap
overlap_pairs = df1_pairs & df2_pairs
overlap_pct = len(overlap_pairs) / len(df2_pairs) * 100

# Filter df1 to only the overlapping (Id, Hour) entries
df1_filtered = df1[df1[['Id', 'ActivityHour']].apply(tuple, axis=1).isin(overlap_pairs)]

# Group by Id and hour, count how many minutes exist in each group
minute_counts = df1_filtered.groupby(['Id', 'ActivityHour']).size().reset_index(name='MinuteCount')

# What % of those have exactly 60 minutes?
complete_hours_pct = (minute_counts['MinuteCount'] == 60).mean() * 100

# Report
print("df1 ActivityMinute range:", df1['ActivityMinute'].min(), "to", df1['ActivityMinute'].max())
print("df2 ActivityHour range:", df2['ActivityHour'].min(), "to", df2['ActivityHour'].max())
print(f"\n➡️ Overlap of (Id, Hour) between df1 and df2: {overlap_pct:.2f}%")
print(f"✅ Of overlapping hours, % that have exactly 60 minutes in df1: {complete_hours_pct:.2f}%")
```

```python
# Convert Date to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Sort data for each user by date
df = df.sort_values(by=['Id', 'Date'])
```

*Calculate Weight Change for Each User*

```python
# Group by user and get first & last weight entries
weight_change = df.groupby('Id').agg(
    start_weight=('WeightKg', 'first'),
    end_weight=('WeightKg', 'last'),
    start_bmi=('BMI', 'first'),
    end_bmi=('BMI', 'last')
)

# Add columns for absolute and percentage weight change
weight_change['abs_weight_change'] = weight_change['end_weight'] - weight_change['start_weight']
weight_change['pct_weight_change'] = (weight_change['abs_weight_change'] / weight_change['start_weight']) * 100
```

1. ✅ % of users whose weight decreased

```python
users_weight_decreased = (weight_change['abs_weight_change'] < 0).sum()
total_users = weight_change.shape[0]
percent_decreased = (users_weight_decreased / total_users) * 100
print(f"🔻 % of users whose weight decreased: {percent_decreased:.2f}%")
```

### **METs**

One metabolic equivalent (MET) is a number that indicates the relative rate at which you burn calories during an activity. If an activity has a value of 4 METs it means that it requires four times the energy than is required by sitting still.
- For example, sitting quietly represents one MET, whereas jogging at six miles per hour represents 10 METs. Multiplying the MET value by the activity duration (in minutes) results in a total (in MET. min). Healthy adults should expend between 450 and 750 MET.

```python
df = pd.read_csv("preprocessed_data\minuteMETsNarrow_merged.csv")
df.head()
```

```python
df['Id'].nunique()
```

🧮 ***Step 1: Calculate Total Energy Spent Per User Per Day***


🧠 Formula:

Energy (MET-minutes) = METs × duration (in minutes)

Since each row represents 1 minute, total energy per row = METs × 1 = just METs.

```python
import pandas as pd

# Convert time column to datetime
df['ActivityMinute'] = pd.to_datetime(df['ActivityMinute'])

# Extract date only (drop time)
df['Date'] = df['ActivityMinute'].dt.date

# Energy per minute = METs (since duration is 1 minute)
# Group by user and date, then sum METs
daily_energy = df.groupby(['Id', 'Date'])['METs'].sum().reset_index(name='Total_METs')

print(daily_energy.head())
```

```python
daily_energy
```

🔎 Step 2: Estimate Energy on Day 1, Day 15, and Last Day

We’ll take:

- Day 1 = first date in dataset

- Day 15 = 15th distinct day per user

- Last Day = last date in dataset

```python
# Sort by user and date
daily_energy = daily_energy.sort_values(['Id', 'Date'])

# Get day counts
daily_energy['Day_Index'] = daily_energy.groupby('Id').cumcount() + 1

# Filter for Day 1, Day 15, and Last Day for each user
day_1 = daily_energy[daily_energy['Day_Index'] == 1]
day_15 = daily_energy[daily_energy['Day_Index'] == 15]
last_day = daily_energy.groupby('Id').tail(1)

# Compare average energy for each
avg_day_1 = day_1['Total_METs'].mean()
avg_day_15 = day_15['Total_METs'].mean()
avg_last_day = last_day['Total_METs'].mean()

print(f"Avg Energy on Day 1: {avg_day_1:.2f}")
print(f"Avg Energy on Day 15: {avg_day_15:.2f}")
print(f"Avg Energy on Last Day: {avg_last_day:.2f}")
```

📊 Step 3: Average Daily Energy Across All Users (and Plot)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Daily average METs across all users
avg_energy_over_time = daily_energy.groupby('Date')['Total_METs'].mean().reset_index()

# Plot
plt.figure(figsize=(12, 5))
sns.lineplot(data=avg_energy_over_time, x='Date', y='Total_METs')
plt.title("Average Daily Energy Spent (MET-minutes) by All Users")
plt.xlabel("Date")
plt.ylabel("Avg Energy (MET-minutes)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Print overall mean
overall_avg = avg_energy_over_time['Total_METs'].mean()
print(f"📌 Average energy spent per day (all users): {overall_avg:.2f} MET-minutes")
```

✅ Here's How to Create the Heatmap (Energy Spent per User per Day)

```python
# Create a mapping from user IDs to 'User 1', 'User 2', ...
unique_ids = daily_energy['Id'].unique()
user_labels = {uid: f'User {i+1}' for i, uid in enumerate(unique_ids)}

# Apply the mapping
daily_energy['UserLabel'] = daily_energy['Id'].map(user_labels)

# Pivot using new user labels
heatmap_data = daily_energy.pivot(index='UserLabel', columns='Date', values='Total_METs')

# Plot heatmap
plt.figure(figsize=(16, 8))
sns.heatmap(heatmap_data, cmap="YlOrRd", linewidths=0.5, linecolor='gray', cbar_kws={'label': 'MET-minutes'})
plt.title("🔥 Heatmap of Energy Spent by Users Over Time")
plt.xlabel("Date")
plt.ylabel("Users")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

✅ Part 2: Percentage of Users Who Increased Their Calories Burnt Over One Month

Logic:

- We’ll check if each user’s average calories burnt in the last week is higher than their first week.


```python
# Ensure datetime type
df['ActivityDay'] = pd.to_datetime(df['ActivityDay'])

# Get first and last 7 days from the full date range
start_date = df['ActivityDay'].min()
end_date = df['ActivityDay'].max()

mid_date = start_date + pd.Timedelta(days=7)
end_range_start = end_date - pd.Timedelta(days=7)

# Calculate average calories per user in first and last 7-day periods
first_week = df[df['ActivityDay'] <= mid_date].groupby('Id')['Calories'].mean()
last_week = df[df['ActivityDay'] >= end_range_start].groupby('Id')['Calories'].mean()

# Align both Series
combined = pd.DataFrame({'start_avg': first_week, 'end_avg': last_week}).dropna()

# Compare and calculate %
combined['increased'] = combined['end_avg'] > combined['start_avg']
percent_increased = 100 * combined['increased'].sum() / len(combined)

print(f"📈 Percentage of users who increased calorie burn: {percent_increased:.2f}%")
```

This looks good; could be a good selling point. There is drop in the avg calories burnt, so try comparing first 7 days of data to first 7 days of May

📌 3. Compare Each User’s Avg Steps: 1st Week of Dataset vs 1st Week of May

```python
import pandas as pd

# Convert date and extract features
df['ActivityDay'] = pd.to_datetime(df['ActivityDay'])
df['Week'] = df['ActivityDay'].dt.isocalendar().week
df['Month'] = df['ActivityDay'].dt.month
df['Year'] = df['ActivityDay'].dt.year

# Get first week of dataset
first_date = df['ActivityDay'].min()
first_week_num = first_date.isocalendar()[1]
first_week_year = first_date.year

# Filter first week of dataset
week1_df = df[(df['Week'] == first_week_num) & (df['Year'] == first_week_year)]
week1_avg = week1_df.groupby('Id')['StepTotal'].mean().reset_index(name='Week1_Avg')

# Filter first week of May (typically ISO Week 18)
may_week1_df = df[(df['Month'] == 5) & (df['Week'] == 18)]
may_week1_avg = may_week1_df.groupby('Id')['StepTotal'].mean().reset_index(name='MayWeek1_Avg')

# Merge both weeks for comparison
compare_may = pd.merge(week1_avg, may_week1_avg, on='Id')

# Calculate percentage change
compare_may['pct_change'] = ((compare_may['MayWeek1_Avg'] - compare_may['Week1_Avg']) / compare_may['Week1_Avg']) * 100

# Compute required stats
total_users = compare_may.shape[0]
users_improved = (compare_may['pct_change'] > 0).sum()
percent_improved = (users_improved / total_users) * 100
max_increase_pct = compare_may['pct_change'].max()

# Show results
print(f"✅ Percentage of users who increased their average daily steps: {percent_improved:.2f}%")
print(f"🚀 Maximum increase in average daily steps: {max_increase_pct:.2f}%")
```
